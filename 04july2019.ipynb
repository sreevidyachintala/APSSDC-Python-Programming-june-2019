```python
#Linear graph

import matplotlib.pyplot as plt

plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])

plt.show()  # Displays the line
```

```python
# Graph To plot distance and time
#import matplotlib.pyplot as plt
distance=list(range(2,11,2))
time=list(range(1,6))
plt.title("Distance vs Time")
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
#v=a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i] * i for i in range(len(t))]

plt.plot(t,v,'go')
plt.grid()
plt.show()
```

```python
#Graph for numbers vs cubes
n=list(range(10))
sqn=[i**3 for i in n]
plt.plot(n,sqn)
#plt.grid()
plt.show()
```

```python
#  Sign Graph
import numpy as np
t= [i/10 for i in range(1,21)]
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title("Time vs Voltage Graph")
plt.plot(t,v)
plt.grid()

plt.xlabel('Time')
plt.ylabel('Voltage')

plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
#Plot income  af Arkanas from 2005 to 2013
import pandas as pd
df=pd.read_csv('DataFiles/Income.csv')
years=df.columns[2:]
#year=list(range(2005,2014))
incomeArkanas=df.values[3,2:]
plt.title("Income of Arkanas")
plt.xlabel("Year")
plt.ylabel("Income")
plt.plot(years,incomeArkanas)
plt.grid()

plt.savefig("Images/Arkanas.png")
plt.show()
#incomeArkanas
```

```python
# Income of all states from 2015 to 2013

df
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates,'go')
plt.title("All states income")
plt.xlabel("Year")
plt.ylabel("Income")
plt.grid()
plt.savefig("Images/AllStates.png")
plt.show()


```
